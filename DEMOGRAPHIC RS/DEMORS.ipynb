```python
def recommend_games(age, gender):
    recommendations = []

    for game in games:
        if age >= game["age_rating"]:
            if game["popular_gender"].lower() == gender.lower() or game["popular_gender"].lower() == "all":
                recommendations.append(game["title"])

    return recommendations
```

```python
nama = input("Masukkan Nama Anda : ")
usia_str = input("Masukkan Usia Anda : ")
usia = int(usia_str)
gender = input("Jenis Kelamin (Male/Female): ")

rekomendasi_lengkap = recommend_games(usia, gender.lower())
top_3_rekomendasi = rekomendasi_lengkap[:3]

if top_3_rekomendasi:
    print(f"Game yang cocok untuk {nama} adalah:")
    for i, game in enumerate(top_3_rekomendasi):
        print(f"{i+1}. {game}")
else:
    print(f"Tidak ada game yang direkomendasikan untuk {nama} dengan usia {usia} dan jenis kelamin {gender}.")
```
